# Ulams Method

Code to test the implementation of Ulam's method for estimating the resonances of the tent map.

```python
import matplotlib.pyplot as plt
import numpy as np

from chebyshev_hofbauer_resonances.linear_tent_map.ulams_method import ulams_method
```

```python
alpha = 1.9


def f(x):
    return np.minimum(alpha * x, alpha * (1 - x))
```

```python
N = 1000  # Number of bins
M = 10000  # Number of random points per bin

L = ulams_method(N, M, f)

eigenvalues, eigenvectors = np.linalg.eig(L)

idx = np.argsort(-np.abs(eigenvalues))
eigenvalues = eigenvalues[idx]
```

```python
plt.scatter(eigenvalues.real, eigenvalues.imag, marker="x")
```
